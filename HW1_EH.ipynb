```python
import random
from keras.models import Sequential
from keras.layers import Dense
from keras import backend as K
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

random.seed(1234)
```

#### 1. Set your random seed to 1234 
#### 2. Load the Fashion-MNIST dataset

```python
# load fashion_mnist data
from keras.datasets import fashion_mnist
(trainX, trainY), (testX, testY) = fashion_mnist.load_data()

# reshape the input matrix
trainX = trainX.reshape((trainX.shape[0], 28 * 28))
testX = testX.reshape((testX.shape[0], 28 * 28))

# scale data to the range of [0, 1]
trainX = trainX.astype("float32") / 255.0
testX = testX.astype("float32") / 255.0
 
# one-hot encode the training and testing labels
trainY = np_utils.to_categorical(trainY, 10)
testY = np_utils.to_categorical(testY, 10)
```

#### 3. Implement a series of neural network models 

i. Initial test

```python
model_ini = Sequential()
model_ini.add(Dense(512, input_shape= (28 * 28,), activation='relu'))
model_ini.add(Dense(512, activation='relu'))
model_ini.add(Dense(512, activation='relu'))
model_ini.add(Dense(512, activation='relu'))
model_ini.add(Dense(10, activation='softmax'))
model_ini.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
result = model_ini.fit(trainX, trainY, validation_split=0.16666, batch_size=512, epochs=200)
```

```python
plt.plot(result.history["val_loss"], label="val_loss")
plt.plot(result.history["val_acc"], label="val_acc")
plt.plot(result.history["loss"], label="train_loss")
plt.plot(result.history["acc"], label="train_acc")
plt.title("Training/validation set accuracy and loss over the epochs")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend()
plt.show()
```

The model starts to degrade when the number of epoch is around 11 or 12.

ii. Implement dropout

```python
from keras.layers import Dropout
model_dropout = Sequential()
model_dropout.add(Dense(512, activation='relu', input_shape= (28 * 28,)))
model_dropout.add(Dropout(0.5))
model_dropout.add(Dense(512, activation='relu'))
model_dropout.add(Dropout(0.5))
model_dropout.add(Dense(512, activation='relu'))
model_dropout.add(Dropout(0.5))
model_dropout.add(Dense(512, activation='relu'))
model_dropout.add(Dropout(0.5))
model_dropout.add(Dense(10, activation='softmax'))
model_dropout.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
result_dropout = model_dropout.fit(trainX, trainY, validation_split=0.16666, batch_size=512, epochs=200)
```

```python
plt.plot(result_dropout.history["val_loss"], label="val_loss_dropout")
plt.plot(result_dropout.history["val_acc"], label="val_acc_dropout")
plt.plot(result_dropout.history["loss"], label="train_loss_dropout")
plt.plot(result_dropout.history["acc"], label="train_acc_dropout")
plt.title("Training/validation set accuracy and loss over the epochs")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend()
plt.show()
```

```python
plt.plot(result.history["val_loss"], label="val_loss")
plt.plot(result.history["val_acc"], label="val_acc")
plt.plot(result_dropout.history["val_loss"], label="val_loss_dropout")
plt.plot(result_dropout.history["val_acc"], label="val_acc_dropout")
plt.title("Comparing initial and dropout validation set accuracy and loss over the epochs")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend()
plt.show()
```

This model performs similarly as the old one in terms of val_acc. However, there is a big improvement of val_loss for the new model, the one with dropout layers, compared to the initial model.
